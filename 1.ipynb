```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)
factorial(5)
```

```python
def factr(num):
    fact = 1
    if(num ==0):
        print("Factorial:"+fact)
    else:
        while( num>0):
            fact *=num
            num -=1
        print("Factorial:", fact)
        
factr(5)
```

```python
# Function parameter types
"""
1. Required arguments
2. Keyword arguments
3. Default arguments
4. Variable-length arguments
"""

# Required arguments
def ex_reqarg(name,age):
    return f"Name: {name}, Age: {age}"

print(ex_reqarg("Alice",30))
print(ex_reqarg(24,"Bob"))

# keyword arguments
print(ex_reqarg(age=28, name="Charlie"))


# Default arguments
def ex_defarg(name, age=25): # If value is not passed, default value set is used
    return f"Name: {name}, Age: {age}"  
print(ex_defarg("David"))
print(ex_defarg("Eve", 32))



# Variable-length arguments
def ex_vararg(*args):
    return args
print(ex_vararg(1,2,3))
print(ex_vararg("a","b","c","d"))
print(ex_vararg(1,"a",True,3.14))
```

```python
# Global and local variables
# All the variables defined outside the functions are global variables and all the 
# variables defined inside the functions are local variables.

c = 7
def my_function():
    c = 5
    print(c)
my_function()
print(c)



d = 7
def my_func():
    global d
    d = 5
    print(d)
my_func()
print(d)
```

```python
def is_palindrome(s):
    str = ''
    for i in s:
        if i.isalnum():
            str += i.lower()
            
            
    return str == str[::-1]
print(is_palindrome("A man, a plan, a canal: Panama"))
```

```python
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    print(num1/num2)
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except ValueError:
    print("Error: Invalid input. Please enter numeric values.")
print("End of the program.")
```

```python
# multiple exception in a single block
try:
    n1 = int(input("Enter first number: "))
    n2 = int(input("Enter second number: "))
    result = n1 / n2
    print("Result:", result)
except (ValueError, ZeroDivisionError) as e:
    print("An error occurred:", e)
except Exception as e:
    print("An unexpected error occurred:", e)
```

```python
# try except else and try except finally
try:
    n1 = int(input("Enter first number: "))
    n2 = int(input("Enter second number: "))
    result = n1 / n2
except (ValueError, ZeroDivisionError) as e:
    print("An error occurred:", e)
else:
    print("The result is:", result)
finally:
    print("Execution completed.")
```
