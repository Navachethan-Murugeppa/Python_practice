```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)
factorial(5)
```

```python
def factr(num):
    fact = 1
    if(num ==0):
        print("Factorial:"+fact)
    else:
        while( num>0):
            fact *=num
            num -=1
        print("Factorial:", fact)
        
factr(5)
```

```python
# Function parameter types
"""
1. Required arguments
2. Keyword arguments
3. Default arguments
4. Variable-length arguments
"""

# Required arguments
def ex_reqarg(name,age):
    return f"Name: {name}, Age: {age}"

print(ex_reqarg("Alice",30))
print(ex_reqarg(24,"Bob"))

# keyword arguments
print(ex_reqarg(age=28, name="Charlie"))


# Default arguments
def ex_defarg(name, age=25): # If value is not passed, default value set is used
    return f"Name: {name}, Age: {age}"  
print(ex_defarg("David"))
print(ex_defarg("Eve", 32))



# Variable-length arguments
def ex_vararg(*args):
    return args
print(ex_vararg(1,2,3))
print(ex_vararg("a","b","c","d"))
print(ex_vararg(1,"a",True,3.14))
```

```python
# python classes and objects
"""
__init__ method works as a constructor in python. It is used to initialize the attributes of the class."""

class Car:
    # these are object attributes initialized using __init__ method
    def __init__(self, model=None, color=None, year=None, speed_limit=None):
        self.model = model
        self.color = color
        self.year = year
        self.speed_limit = speed_limit
    
    def display_info(self):
        print(f"MOdel ;{self.model}, Color: {self.color}, Year: {self.year}, Speed Limit: {self.speed_limit} Km/hr")

car4 = Car("BMW X5", "Black", 2021, 250)
car4.display_info()
car5 = Car("Mercedes GLE")
car5.display_info()
```

```python
#different methods in python classes -> Regular methods, class methods and static methods
import datetime
class Employee:
    increase_amount = 1.05  # class variable
    def __init__(self, first_name, last_name, salary, skills):
        self.first_name = first_name
        self.last_name = last_name
        self.salary = salary
        self.skills = skills
        self.email = f"{first_name.lower()}.{last_name.lower()}@company.com"
    
    def display_info(self):
        print("======= Employee Information =======")
        print(f"Name: {self.first_name} {self.last_name}, Salary: {self.salary}, Skills: {', '.join(self.skills)}, Email: {self.email}")
    
    def inc_salary(self, percent=5):
        self.salary += self.salary * (percent / 100)
        print(f"Salary increased by {percent}%. New Salary: {self.salary}")
    
    def add_skill(self, skill):
        self.skills.append(skill)
        print(f"Skill '{skill}' added. Updated Skills: {', '.join(self.skills)}")

    @classmethod
    def set_increase_amt(cls, amount):
        # pass
        cls.increase_amount = amount
        
    @staticmethod
    def is_workday(day):
        if day.weekday() in (5, 6):
            return False
        return True


emp1 = Employee("John", "Doe", 60000, ["Python", "Java", "SQL"])
emp1.display_info()
```

```python
# Advanced lists operations
# append(), extend(), insert(), remove(), pop(), clear(), index(), count(), sort(), reverse(), copy()

# Advanced list operations
# append(), extend(), insert(), remove(), pop(), clear(), index(), count(), sort(), reverse(), copy()

# ========================================
# 1. append() - Add single element at the end
# ========================================
fruits = ['apple', 'banana', 'cherry']
print("Original list:", fruits)

fruits.append('orange')
print("After append('orange'):", fruits)

fruits.append(['mango', 'grape'])  # Adds the entire list as one element
print("After append(['mango', 'grape']):", fruits)
print()

# ========================================
# 2. extend() - Add multiple elements at the end
# ========================================
colors = ['red', 'green', 'blue']
print("Original list:", colors)

colors.extend(['yellow', 'purple'])
print("After extend(['yellow', 'purple']):", colors)

colors.extend('pink')  # Adds each character separately
print("After extend('pink'):", colors)
print()

# ========================================
# 3. insert() - Add element at specific position
# ========================================
numbers = [1, 2, 3, 5, 6]
print("Original list:", numbers)

numbers.insert(3, 4)  # Insert 4 at index 3
print("After insert(3, 4):", numbers)

numbers.insert(0, 0)  # Insert at beginning
print("After insert(0, 0):", numbers)

numbers.insert(len(numbers), 7)  # Insert at end (same as append)
print("After insert(len(numbers), 7):", numbers)
print()

# ========================================
# 4. remove() - Remove first occurrence of value
# ========================================
animals = ['cat', 'dog', 'rabbit', 'dog', 'hamster']
print("Original list:", animals)

animals.remove('dog')  # Removes first 'dog'
print("After remove('dog'):", animals)

# animals.remove('lion')  # This would raise ValueError if uncommented
print()

# ========================================
# 5. pop() - Remove and return element at index
# ========================================
items = ['pen', 'pencil', 'eraser', 'ruler', 'notebook']
print("Original list:", items)

removed_item = items.pop()  # Remove last item
print(f"Popped item (no index): {removed_item}")
print("List after pop():", items)

removed_item = items.pop(1)  # Remove item at index 1
print(f"Popped item at index 1: {removed_item}")
print("List after pop(1):", items)
print()

# ========================================
# 6. clear() - Remove all elements
# ========================================
temp_list = [1, 2, 3, 4, 5]
print("Original list:", temp_list)

temp_list.clear()
print("After clear():", temp_list)
print()

# ========================================
# 7. index() - Find index of first occurrence
# ========================================
letters = ['a', 'b', 'c', 'd', 'b', 'e']
print("Original list:", letters)

idx = letters.index('c')
print(f"Index of 'c': {idx}")

idx = letters.index('b')  # Returns first occurrence
print(f"Index of 'b' (first occurrence): {idx}")

idx = letters.index('b', 2)  # Start searching from index 2
print(f"Index of 'b' starting from index 2: {idx}")

# letters.index('z')  # This would raise ValueError if uncommented
print()

# ========================================
# 8. count() - Count occurrences of value
# ========================================
values = [1, 2, 3, 2, 4, 2, 5, 2, 6]
print("Original list:", values)

count_2 = values.count(2)
print(f"Count of 2: {count_2}")

count_7 = values.count(7)
print(f"Count of 7: {count_7}")
print()

# ========================================
# 9. sort() - Sort list in place (ascending)
# ========================================
nums = [5, 2, 8, 1, 9, 3, 7]
print("Original list:", nums)

nums.sort()
print("After sort():", nums)

nums.sort(reverse=True)
print("After sort(reverse=True):", nums)

# Sort strings
words = ['banana', 'apple', 'cherry', 'date']
print("\nOriginal words:", words)
words.sort()
print("After sort():", words)

words.sort(key=len)  # Sort by length
print("After sort(key=len):", words)
print()

# ========================================
# 10. reverse() - Reverse list in place
# ========================================
sequence = [10, 20, 30, 40, 50]
print("Original list:", sequence)

sequence.reverse()
print("After reverse():", sequence)
print()

# ========================================
# 11. copy() - Create shallow copy of list
# ========================================
original = [1, 2, 3, 4, 5]
print("Original list:", original)

# Shallow copy using copy()
copied = original.copy()
print("Copied list:", copied)

copied.append(6)
print("After appending 6 to copied list:")
print("Original:", original)
print("Copied:", copied)
print()

# Comparison with assignment
list1 = [1, 2, 3]
list2 = list1  # Not a copy, just a reference
list3 = list1.copy()  # Actual copy

list2.append(4)
list3.append(5)

print("list1:", list1)
print("list2 (reference):", list2)
print("list3 (copy):", list3)
print()
```

```python
# ========================================
# COMBINED OPERATIONS - Real World Examples
# ========================================

# Example 1: Shopping Cart Management
print("="*50)
print("Example 1: Shopping Cart")
print("="*50)

cart = []
print("Empty cart:", cart)

# Add items
cart.append('Laptop')
cart.append('Mouse')
cart.extend(['Keyboard', 'Monitor'])
print("After adding items:", cart)

# Insert item at specific position
cart.insert(1, 'Headphones')
print("After inserting Headphones at index 1:", cart)

# Remove item
cart.remove('Mouse')
print("After removing Mouse:", cart)

# Check if item exists
if 'Laptop' in cart:
    print(f"Laptop is at index: {cart.index('Laptop')}")

print(f"Total items in cart: {len(cart)}")
print()

# Example 2: Grade Management System
print("="*50)
print("Example 2: Grade Management")
print("="*50)

grades = [85, 92, 78, 90, 88, 92, 76, 95]
print("Original grades:", grades)

# Count specific grade
count_92 = grades.count(92)
print(f"Number of students with 92: {count_92}")

# Sort grades
sorted_grades = grades.copy()
sorted_grades.sort(reverse=True)
print("Sorted grades (descending):", sorted_grades)

# Find highest and lowest
print(f"Highest grade: {sorted_grades[0]}")
print(f"Lowest grade: {sorted_grades[-1]}")

# Remove lowest grade
grades_copy = grades.copy()
grades_copy.remove(min(grades_copy))
print("After removing lowest grade:", grades_copy)
print()

# Example 3: Task Manager
print("="*50)
print("Example 3: Task Manager")
print("="*50)

tasks = ['Write report', 'Send emails', 'Meeting at 2pm']
print("Current tasks:", tasks)

# Add new tasks
tasks.insert(0, 'Review code')  # Highest priority
tasks.append('Call client')  # Lowest priority
print("After adding new tasks:", tasks)

# Complete a task
completed = tasks.pop(0)
print(f"Completed task: {completed}")
print("Remaining tasks:", tasks)

# Reorder tasks
tasks.reverse()
print("Tasks reversed:", tasks)
print()

# Example 4: Playlist Management
print("="*50)
print("Example 4: Playlist Management")
print("="*50)

playlist = ['Song A', 'Song B', 'Song C', 'Song D']
print("Original playlist:", playlist)

# Add songs
playlist.extend(['Song E', 'Song F'])
print("After adding songs:", playlist)

# Move song to top
favorite = 'Song D'
if favorite in playlist:
    playlist.remove(favorite)
    playlist.insert(0, favorite)
print(f"After moving {favorite} to top:", playlist)

# Shuffle simulation (reverse for simple example)
shuffled = playlist.copy()
shuffled.reverse()
print("Shuffled playlist:", shuffled)
print("Original playlist (unchanged):", playlist)
print()

# Example 5: Inventory Management
print("="*50)
print("Example 5: Inventory Management")
print("="*50)

inventory = ['item1', 'item2', 'item3', 'item2', 'item4', 'item2']
print("Current inventory:", inventory)

# Count specific item
item_count = inventory.count('item2')
print(f"'item2' count: {item_count}")

# Remove all occurrences of an item
while 'item2' in inventory:
    inventory.remove('item2')
print("After removing all 'item2':", inventory)

# Restock items
inventory.extend(['item5', 'item6', 'item7'])
inventory.sort()
print("After restocking and sorting:", inventory)
print()
```

```python
# ========================================
# QUICK REFERENCE - All Methods Together
# ========================================

print("="*60)
print("COMPREHENSIVE LIST METHODS DEMONSTRATION")
print("="*60)

# Starting list
my_list = [3, 1, 4, 1, 5, 9, 2, 6]
print(f"Starting list: {my_list}\n")

# 1. append()
my_list.append(7)
print(f"1. append(7):           {my_list}")

# 2. extend()
my_list.extend([8, 9])
print(f"2. extend([8, 9]):      {my_list}")

# 3. insert()
my_list.insert(0, 0)
print(f"3. insert(0, 0):        {my_list}")

# 4. index()
idx = my_list.index(5)
print(f"4. index(5):            Position {idx}")

# 5. count()
cnt = my_list.count(1)
print(f"5. count(1):            {cnt} occurrences")

# 6. remove()
my_list.remove(9)  # Removes first 9
print(f"6. remove(9):           {my_list}")

# 7. pop()
popped = my_list.pop()
print(f"7. pop():               {my_list} (removed {popped})")

# 8. copy()
list_copy = my_list.copy()
print(f"8. copy():              {list_copy} (copied)")

# 9. sort()
my_list.sort()
print(f"9. sort():              {my_list}")

# 10. reverse()
my_list.reverse()
print(f"10. reverse():          {my_list}")

# 11. clear()
temp = my_list.copy()
temp.clear()
print(f"11. clear() on copy:    {temp}")

print(f"\nFinal list: {my_list}")
print(f"Length: {len(my_list)}")
```
